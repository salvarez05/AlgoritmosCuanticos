```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector, partial_trace, Pauli
import numpy as np
```

```python

qc = QuantumCircuit(3)

qc.rx(np.pi/2, 0)          # X**0.25 â‰ˆ Rx(pi/2)
qc.h(1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.h(0)
qc.cx(1, 2)
qc.cz(0, 2)

simulator = AerSimulator()
qc_t = transpile(qc, simulator)
state = Statevector.from_instruction(qc_t)

msg = QuantumCircuit(1)
msg.rx(np.pi/2, 0)
msg_state = Statevector.from_instruction(msg)

paulis = [Pauli('X'), Pauli('Y'), Pauli('Z')]

bloch_msg = [np.real(msg_state.expectation_value(p)) for p in paulis]

bob = partial_trace(state, [0, 1])
bloch_bob = [np.real(bob.expectation_value(p)) for p in paulis]

print("Bloch vector del mensaje:", np.round(bloch_msg, 3))
print("Bloch vector del qubit de Bob:", np.round(bloch_bob, 3))
```
